```python
# Business impact metrics
impact_metrics = {
    'Time Savings': {
        'Current': '4-6 hours per contract',
        'Phase 1': '2-3 hours per contract',
        'Phase 2': '1-2 hours per contract',
        'Phase 3': '30 minutes per contract',
        'Phase 4': '15 minutes per contract',
        'Phase 5': '5 minutes per contract'
    },
    'Risk Detection': {
        'Current': 'Manual review (70% accuracy)',
        'Phase 1': 'Keyword-based (80% accuracy)',
        'Phase 2': 'ML-based (90% accuracy)',
        'Phase 3': 'RAG-enhanced (92% accuracy)',
        'Phase 4': 'Production-tested (94% accuracy)',
        'Phase 5': 'Enterprise-grade (96% accuracy)'
    },
    'Cost Reduction': {
        'Current': 'Base cost',
        'Phase 1': '25% reduction',
        'Phase 2': '50% reduction',
        'Phase 3': '70% reduction',
        'Phase 4': '80% reduction',
        'Phase 5': '90% reduction'
    }
}

# Create business impact visualization
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for idx, (metric, data) in enumerate(impact_metrics.items()):
    stages = list(data.keys())
    values = list(data.values())
    
    # Convert to numeric where possible 
    numeric_values = []
    for value in values:
        if '%' in value:
            # Extract percentage number from various formats
            if 'reduction' in value:
                # Handle "25% reduction" format
                numeric_values.append(float(value.replace('%', '').replace(' reduction', '')))
            elif 'accuracy' in value:
                # Handle "70% accuracy" format - extract number from parentheses
                import re
                match = re.search(r'\((\d+)%', value)
                if match:
                    numeric_values.append(float(match.group(1)))
                else:
                    numeric_values.append(0)
            else:
                # Handle other percentage formats
                numeric_values.append(float(value.replace('%', '')))
        elif 'hours' in value or 'minutes' in value:
            # Handle time formats
            if 'hours' in value:
                # Extract numbers from range like "4-6" -> average of 4 and 6 = 5
                numbers = value.replace(' hours per contract', '').split('-')
                if len(numbers) == 2:
                    avg_hours = (float(numbers[0]) + float(numbers[1])) / 2
                    numeric_values.append(avg_hours * 60)  # Convert to minutes
                else:
                    numeric_values.append(float(numbers[0]) * 60)
            else:
                # Handle single minutes like "30 minutes"
                minutes = float(value.replace(' minutes per contract', ''))
                numeric_values.append(minutes)
        else:
            # Handle non-numeric values like "Base cost"
            numeric_values.append(0)
    
    bars = axes[idx].bar(range(len(stages)), numeric_values, 
                        color=colors[:len(stages)], alpha=0.7)
    
    # Add text labels
    for i, (stage, value) in enumerate(zip(stages, values)):
        axes[idx].text(i, numeric_values[i] + max(numeric_values)*0.05, value, 
                      ha='center', va='bottom', fontsize=8, rotation=45)
    
    axes[idx].set_title(metric, fontsize=14, fontweight='bold')
    axes[idx].set_xticks(range(len(stages)))
    axes[idx].set_xticklabels(stages, rotation=45, ha='right', fontsize=8)

plt.tight_layout()
plt.show()
```
